### Feature Selection
1. Dropping constant features

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.DataFrame({'A' : [1,2,4,1,2,4],
                     'B' : [4,5,6,7,8,9],
                     'C' : [0,0,0,0,0,0],
                     'D' : [1,1,1,1,1,1]})
```

```python
data.head()
```

* C and D are having constant features so we will try dropping them
* For that we will use Variance Threshold Method

### 1. Varince Threshold Method
* It removes all low variance features.
* This feature selection algorithm looks only at the features (X), not the desired outputs (y), and can thus be used for **unsupervised learning**.
* It removes 0 variance features.

```python
from sklearn.feature_selection import VarianceThreshold
var_thresh = VarianceThreshold(threshold=0)
var_thresh.fit(data)
```

```python
var_thresh.get_support()
```

* False indicates 0 or low variance

```python
constant_columns = [column for column in data.columns
                    if column not in data.columns[var_thresh.get_support()]]

print(constant_columns)
```

```python
for i in constant_columns:
    print(i)
```

```python
data.drop(constant_columns,axis = 1)
```

### Lets Practice on bigger datasets

```python
df = pd.read_csv('santander.csv')
```

```python
df.head()
```

```python
# seperating input and output
X = df.drop(['TARGET'], axis = 1)
y = df['TARGET']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state= 0)
```

### lets apply variance threshold on this X_train as X_train is the input.

```python
X_train.shape
```

```python
var_th = VarianceThreshold(threshold=0)
var_th.fit(X_train)
```

```python
var_th.get_support()
```

```python
sum(var_th.get_support())
```

```python
drop_cols = [column for column in X_train.columns
             if column not in X_train.columns[var_th.get_support()]]

print(len(drop_cols))
```

```python
for i in drop_cols:
    print(i)
```

```python
X_train.drop(drop_cols, axis=1)
```
