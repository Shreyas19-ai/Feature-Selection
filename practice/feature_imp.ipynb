## Feature Importance
### It gives a score for each feature of your data, the higher the score more relevant the feature is

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('mobile_dataset.csv')
df.head()
```

```python
X = df.drop(['price_range'], axis = 1)
y = df['price_range']
```

```python
from sklearn.ensemble import ExtraTreesClassifier
etc = ExtraTreesClassifier()
etc.fit(X, y)
```

```python
print(etc.feature_importances_)
```

```python
rank = pd.Series(etc.feature_importances_, index = X.columns)
```

```python
rank.nlargest(10).plot(kind = 'barh')
plt.show()
```
